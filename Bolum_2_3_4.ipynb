```python
import numpy as np
import skfuzzy as fuzz
import matplotlib.pyplot as plt
```

```python
print("Numpy sürümü: ", np.__version__)
print("Scikit-fuzzy sürümü: ", fuzz.__version__)
print("Matplotlib sürümü: ", plt.matplotlib.__version__)
```

```python
# Bulanık mantık kurulum işlemleri gerçekleştiriliyor. Aşağıdaki kodlarda aşama aşama işlemler bulunmaktadır.
# "0 ile 1 aralığında trapezoidal üyelik fonksiyonu oluştur"
x = np.arange(0, 5.05, 0.1) # [0, 5.05] aralığında 0.1 artan x değerlerine sahip bir dizi oluştur
mfx = fuzz.trapmf(x, [2, 2.5, 3, 4.5]) # x dizisi için üyelik fonksiyonu hesapla

# # Beş farklı yöntemle üyelik fonksiyonunun üyesini hesaplama işlemi gerçekleştirilecektir.
# Bunlar, centroid, bisector, mom, som ve lom 'dur.

defuzz_centroid = fuzz.defuzz(x, mfx, 'centroid')  
defuzz_bisector = fuzz.defuzz(x, mfx, 'bisector')
defuzz_mom = fuzz.defuzz(x, mfx, 'mom')
defuzz_som = fuzz.defuzz(x, mfx, 'som')
defuzz_lom = fuzz.defuzz(x, mfx, 'lom')

# Oluşturulacak Dikey çizgiler için veriler hazırlanıyor
labels = ['centroid', 'bisector', 'mean of maximum', 'min of maximum', 'max of maximum']
xvals = [defuzz_centroid, defuzz_bisector, defuzz_mom, defuzz_som, defuzz_lom]
colors = ['r', 'b', 'g', 'c', 'm']
ymax = [fuzz.interp_membership(x, mfx, i) for i in xvals]

# Üyelik fonksiyonuna karşı üyelik fonksiyonunun üyesini göster ve karşılaştır
plt.figure(figsize=(8, 5))
plt.plot(x, mfx, 'k')  #Üyelik fonksiyonunu siyah çizgiyle çiz
for xv, y, label, color in zip(xvals, ymax, labels, colors):
    plt.vlines(xv, 0, y, label=label, color=color) # Dikey çizgileri ekle
plt.ylabel('Fuzzy membership') # Y ekseninin başlığı
plt.xlabel('Universe variable (arb)')  # X ekseninin başlığı
plt.ylim(-0.1, 1.1) # Y ekseninin sınırları
plt.legend(loc=2)# Etiketleri göster
plt.show() # Sonuç Grafiğini göster
```
